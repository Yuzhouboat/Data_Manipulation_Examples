### 3. Create a new column counting number of preferences subscribers select. If the value of ebb_preference is empty, subscriber will receive all the genres that are available.

```python
# Find all the genres
# genres are seperated with commas (need get rid of all other punctuations beforehand)
ebb = df['ebb_preferences'].dropna().str.lower().str.replace('[^\w,]|_','').str.replace(',',' ')
genre = set(x for l in ebb.str.split() for x in l)

# Check if there are issues
genre
```

```python
# Several Issues here
    # 'null' and 'someebbpref' need to be treated the same as Nan
    # 'biographies' and 'biography' need to be treated the same
    # 'hist' and 'history' need to be treated the same
    # There some genres only showing part of the word, like 'childerns'
        # all genre start with 'c' count as 'childerns'
        # all genre start with 'g' count as 'generalnonfiction'
        # all genre start with 'life' count as 'lifestyleandcooking'
        # all genre start with 'liter' count as 'literaryfiction'
        # all genre start with 'myster' count as 'mysteriesthrillerstruecrime'
        # all genre start with 'p' count as 'poetry'
        # all genre start with 'r' count as 'romance'
        # all genre start with 's' count as 'sciencefictionfantasy'
        # all genre start with 't' count as 'thrillerstruecrime'
        # all genre start with 'y' count as 'youngadult'
    

# Fixing issues
def Fixing(ebb):
  ebb = ebb.str.lower().str.replace('[^\w,]|_','').str.replace(',',' ')
  ebb = ebb.str.replace('someebbpref','').str.replace('biographies','biography').str.replace('history','hist').str.replace('null','')
  ebb = ebb.str.replace(r'\b([c]\w+)\b|\b([c])\b','childrens',regex = True)
  ebb = ebb.str.replace(r'\b([g]\w+)\b|\b([g])\b','generalnonfiction',regex = True)
  ebb = ebb.str.replace(r'\b([life]\w+)\b','lifestyleandcooking',regex = True)
  ebb = ebb.str.replace(r'\b([liter]\w+)\b','literaryfiction',regex = True)
  ebb = ebb.str.replace(r'\b([myster]\w+)\b','mysteriesthrillerstruecrime',regex = True)
  ebb = ebb.str.replace(r'\b([p]\w+)\b|\b([p])\b','poetry',regex = True)
  ebb = ebb.str.replace(r'\b([r]\w+)\b|\b([r])\b','romance',regex = True)
  ebb = ebb.str.replace(r'\b([s]\w+)\b|\b([s])\b','sciencefictionfantasy',regex = True)
  ebb = ebb.str.replace(r'\b([t]\w+)\b|\b([t])\b','thrillerstruecrime',regex = True)
  ebb = ebb.str.replace(r'\b([y]\w+)\b|\b([y])\b','youngadult',regex = True)
  ebb = ebb.str.split()
  return ebb

ebb = df['ebb_preferences'].dropna()
ebb = Fixing(ebb)
genre = set(x for l in ebb for x in l)

# Check Again
genre
```

```python
# Alter the DataFrame
genre_str = str(genre)
# If the value of ebb_preference is empty, subscriber will receive all the genres that are available.
ebb = df['ebb_preferences'].fillna(genre_str)
df['ebb_preferences'] = Fixing(ebb)
df['ebb_preferences'] = df['ebb_preferences'].apply(np.array)
```

### Data Preprocessing

```python
# Remove Features that obviously cannot contribute to forecasting
# Id is unique
# Names (First and Last) not helping
# Year,Month is derived from Signup Date
# ebb_preferences already breaked down into dummies
# Mail Domain not usefull
# Geolocation only use Country
#df = df_original.copy()
df.drop(['Profile.Id','Year','Month', 'first_name','last_name','ebb_preferences','Domain', 'Geolocation.City','Geolocation.State','Geolocation.Zip'], axis = 1, inplace = True)
df.dtypes
```

```python
# Age and Gender has over 99% Na values drop them
df.drop(['age','gender'], axis = 1, inplace = True)
```

```python
# Mange the datetime variables
# For the forecasting purpose round to day
df.Signup = df.Signup.dt.date
df['Optout.Time'] = pd.to_datetime(df['Optout.Time']).dt.date
df['Last.Open'] = pd.to_datetime(df['Last.Open']).dt.date
df['Last.Click'] = pd.to_datetime(df['Last.Click']).dt.date
```

```python
#Covert Last Clicks and Last Open
#The duration between Last Clicks (Last Opens) and Sign up date makes more sense here.
df['Last.Open'] = (df['Last.Open'] - df['Signup']).dt.days
df['Last.Click'] = (df['Last.Click'] - df['Signup']).dt.days
# fill na using means
df['Last.Click'].fillna(0,inplace = True)
df['Last.Open'].fillna(0,inplace = True)
```

```python
df['Top.Device'].isnull().mean()
df['Top.Device'].value_counts()
```

```python
# Make All categories below Safari into other categories, Also Make Na value its own category
df['Top.Device'].fillna('Notknowdevice', inplace = True)
cate_keep = list(df['Top.Device'].value_counts().index[:10])+['Notknowdevice']
df['Top.Device'][~df['Top.Device'].isin(cate_keep)] = 'Otherdevice'
```

```python
df['referral_source'].value_counts()
```

```python
# Keep the first 12 categories (till google) group others into other
cate_keep = list(df['referral_source'].value_counts().index[:12])
df['referral_source'][~df['referral_source'].isin(cate_keep)] = 'Othersource'
```

```python
#Country divided into US and non-US, and keep NA in its own category
df['Geolocation.Country'].fillna('Notknowcountry', inplace = True)
cate_keep = ['US','Notknowcountry']
df['Geolocation.Country'][~df['Geolocation.Country'].isin(cate_keep)] = 'Non-US'
```

### The Forecasting Model for Customer optout or not

To find when a customer is going to optout the newsleeter, we need first have a model to detect if the customer is going to churn or not. Then in the group of customers that have high possibility to chunk, we then use another model to determine when that churn will happen.

Therefore, in the first model to determine if it is going to churn or not, we use 'Engagement' to drive our target variable. And the 'Optout.Time' is not need in this model train.

```python
Churn_df = df.drop(['Optout.Time', 'Signup'],axis =1)
```

```python
Churn_df.Engagement = (Churn_df.Engagement == 'optout').astype(int)
```

```python
Churn_df.head()
```

**Above is the data for the churn detection model. The requirment is forecast when customers is going to optout, so skip the model here.**

### The prediction model for when a customers is going to optout

```python
#Here we do not need the 'Engagement', because the training dataset now only contains customers that already optout.
Predict_df = df[df['Engagement'] == 'optout'].drop(['Engagement'],axis =1)
#The target variable here is the duration of time 
Predict_df['Optout.Time'] = (Predict_df['Optout.Time'] - Predict_df['Signup']).dt.days
```

```python
Predict_df.dropna(subset=['Optout.Time'], inplace = True)
Predict_df.drop(['Signup'],axis = 1,inplace = True)
```

```python
Predict_df.head()
```

**Try a Random forest model on the dataset**

```python
#Covert categorical into dummies
country = pd.get_dummies(Predict_df['Geolocation.Country'])
referral = pd.get_dummies(Predict_df['referral_source'])
device = pd.get_dummies(Predict_df['Top.Device'])
RF_df = Predict_df.join([country,referral, device],how='inner').drop(['Geolocation.Country','referral_source','Top.Device'],axis = 1)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import ShuffleSplit
y = RF_df['Optout.Time']
X = RF_df.drop(['Optout.Time'],axis = 1)
forest = RandomForestRegressor(n_estimators=50,max_features=2,max_depth = 10)
cvf = ShuffleSplit(n_splits=25, test_size=0.25)
scores = cross_val_score(forest, X, y, cv=cvf)
print(np.mean(scores))
```

**Cross Validation Score (R square) in test set is about 80%**
